# **Upload Images by running the below cell**

```
from google.colab import files
uploaded = files.upload()
images=[]
for fn in uploaded.keys():
  img = cv2.imread('/content/'+fn)
  img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
  img = cv2.resize(img, (60,60))
  images.append(img)
normalised_images = np.array(images, dtype = np.float32)/255
```

# **Loading the model from the downloaded model architecture and weights**

```
model = keras.models.load_model('/content/my_model_fin.h5')
```

# **Converting the images list into numpy array**

```
X_test = np.expand_dims(normalised_images, axis = -1)
```

# **Prediction of the class of the uploaded images**

```
for image in X_test:
  pred = np.argmax(model.predict(np.reshape(image, (1,60,60,1))))
  plt.figure()
  plt.title(d[pred])
  plt.imshow(image.reshape(60,60), cmap = 'gray')
```
